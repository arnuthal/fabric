```python
# Import
import os
import pandas as pd
from pyspark.sql.types import *
```

```python
# Read Files
folder_path = "/lakehouse/default/Files/RawData"  

files = os.listdir(folder_path)
```

```python
#Define Schema
schema = StructType([
    StructField("Region", StringType()),
    StructField("Country", StringType()),
    StructField("Item Type", StringType()),
    StructField("Sales Channel", StringType()),
    StructField("Order Priority", StringType()),
    StructField("Order Date", StringType()),
    StructField("Order ID", StringType()),
    StructField("Ship Date", StringType()),
    StructField("Units Sold", StringType()),
    StructField("Unit Price", StringType()),
    StructField("Unit Cost", StringType()),
    StructField("Total Cost",StringType()),
    StructField("Total Profit", StringType())
    ])
staging_Table_df = spark.createDataFrame(spark.sparkContext.emptyRDD(), schema)
```

```python
#read files and remove space character from column headers
for file in files:
    df = spark.read.format("csv").option("header","true").schema(schema).load("Files/RawData/" + file)
    staging_Table_df = staging_Table_df.union(df)


def replace_spaces_in_column_names(df):
    new_column_names = [column.replace(" ", "_") if " " in column else column for column in df.columns]
    return df.toDF(*new_column_names)

# Rename the columns
staging_Table_df = replace_spaces_in_column_names(staging_Table_df)
```

```python
# Create table
staging_Table_df.write.format("delta").mode("overwrite").saveAsTable("Staging_sales")
```
